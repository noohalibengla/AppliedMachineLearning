```python
import os
import zipfile
import shutil

# Define paths
base_dir = "/content/birds_dataset"
chicken_zip = "/content/chicken_images.zip"
duck_zip = "/content/duck_images.zip"

# Extract zip files
with zipfile.ZipFile(chicken_zip, 'r') as zip_ref:
    zip_ref.extractall("/content/temp_chicken")

with zipfile.ZipFile(duck_zip, 'r') as zip_ref:
    zip_ref.extractall("/content/temp_duck")

# Create class folders
os.makedirs(os.path.join(base_dir, "chicken"), exist_ok=True)
os.makedirs(os.path.join(base_dir, "duck"), exist_ok=True)

# Move actual images from nested folders
chicken_folder = "/content/temp_chicken/chicken_images"
duck_folder = "/content/temp_duck/duck_images"

# Move chicken images
for file in os.listdir(chicken_folder):
    if file.lower().endswith(('.jpg', '.jpeg', '.png')):
        shutil.move(os.path.join(chicken_folder, file),
                    os.path.join(base_dir, "chicken", file))

# Move duck images
for file in os.listdir(duck_folder):
    if file.lower().endswith(('.jpg', '.jpeg', '.png')):
        shutil.move(os.path.join(duck_folder, file),
                    os.path.join(base_dir, "duck", file))
```

```python
import torch
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, random_split
import os
import numpy as np

data_dir = '/content/birds_dataset'

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

dataset = datasets.ImageFolder(root=data_dir, transform=transform)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16)
```

```python
from torchvision import datasets, transforms, models

model = models.resnet18(pretrained=True)

# Freeze feature extractor
for param in model.parameters():
    param.requires_grad = False

# Modify classifier
num_features = model.fc.in_features
model.fc = torch.nn.Linear(num_features, 2)  # 2 classes: duck and chicken

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

```python
import torch.nn as nn
import torch.optim as optim
import numpy as np

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.fc.parameters(), lr=0.001)

epochs = 5
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {running_loss/len(train_loader):.4f}")
```

```python
from sklearn.metrics import classification_report

model.eval()
y_true, y_pred = [], []

with torch.no_grad():
    for images, labels in val_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, preds = torch.max(outputs, 1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(preds.cpu().numpy())

print(classification_report(y_true, y_pred, target_names=dataset.classes))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def imshow(img, title):
    img = img.numpy().transpose((1, 2, 0))
    img = np.clip(img, 0, 1)
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()

model.eval()
images, labels = next(iter(val_loader))
images, labels = images.to(device), labels.to(device)
outputs = model(images)
_, preds = torch.max(outputs, 1)
```

```python
for i in range(16):
    imshow(images[i].cpu(), f"Predicted: {dataset.classes[preds[i]]}, Actual: {dataset.classes[labels[i]]}")
```
